```python
def calculator(n1,n2):
    print(f'{n1} + {n2} = {n1 + n2}')
    print(f'{n1} - {n2} = {n1 - n2}')
    print(f'{n1} * {n2} = {n1 * n2}')
    print(f'{n1} / {n2} = {n1 / n2}')
calculator(10,20)
```

```python
def calculator(*n):
    print(f'{n} + {n} = {n + n}')
    print(type(n))  # 튜플형태 iterable 객체

calculator(10,20)
```

```python
def calculator(*n):
    for num in n:
        print(num,end='')
    print()
calculator()
calculator(10)
calculator(10,20)
calculator(10,20,30)
```

```python
# 국어, 영어, 수학 입력->  총점, 평균 출력
def score(kor, eng, math):
    sum = kor + eng + math
    avg = sum/ 3
    print(f'총점:{sum}')
    print(f'평균:{round(avg,2)}')

korS = int(input('국어점수입력:'))
engS = int(input('영어점수입력:'))
mathS = int(input('수학점수입력:'))

score(korS,engS,mathS)
```

### 지역변수와 전역변수
#### 전역변수
- 함수 밖에 선언된 변수로 어디에서나 사용은 가능하지만 함수 안에서 수정할 수 없음.
#### 지역변수
- 함수 안에 선언된 변수로 함수 안에서만 사용 가능.
#### global 키워드 
- global을 사용하면 함수 안에서도 전역변수의 값을 수정 가능.
    - 전역변수가 함수 내에서는 수정이 불가능하기에 global 키워드를 사용하여 함수 내에서도 전역변수 수정을 가능케한다.

```python
# 사용자가 가로,세로 길이 입력-> 삼각형과 사각형 넓이 

def area(): 
    recArea = height * width
    triArea = height * width/2
    print(recArea)
    print(triArea)
width = int(input('가로:'))
height = int(input('세로:'))

area()
```

```python
# 방문객 수 카운트하는 함수
totalVisitor = 0 #전역변수

def count():
    global totalVisitor 
    
    #전역변수가 함수 내에서는 수정이 불가능하기에 global 키워드를 사용하여 함수 내에서도 전역변수 수정을 가능케한다.

    totalVisitor += 1
    print(f'누적 방문객 수:{totalVisitor}')
count()
count()
count()
```

### 중첩함수 
- 함수 안에 또 다른 함수가 있는 형태
- 내부 함수를 함수 밖에서 호출할 수 없음



```python
def out_function():
    print('sout_function called!!')
    
    def in_function():
        print('in_function called!!')
    in_function()
out_function()
in_function()
```

```python
# 덧셈, 뺄셈, 곱셈, 나눗셈
def calculator(n1,n2,operator):
    def add():
        print(f'덧셈:{round((n1+n2),2)}')
    def sub():
        print(f'뺄셈:{round((n1-n2),2)}')
    def mul():
        print(f'곱셈:{round((n1*n2),2)}')
    def div():
        print(f'나눗셈:{round((n1/n2),2)}')
    if operator == 1 :
        add()
    elif operator == 2 :
        sub()
    elif operator == 3 :
        mul()
    elif operator == 4 :
        div()

while True:
    num1 = int(input('실수(n1)입력:'))
    num2 = int(input('실수(n2)입력:'))
    operatorNum = int(input('1.덧셈,2.뺄셈,3.곱셈,4.나눗셈,5.종료'))

    if operatorNum == 5:
        print('Bye~~')
        break

    calculator(num1,num2,operatorNum)

```

```python
### lambda 함수
- 함수 선언 간단히!
- lambda 키워드를 이용하면 함수 선언 보다 간단
```

```python
def calculator(n1,n2):
    return n1+n2
calculator(10,20)
```

```python
calculator = lambda n1,n2 : n1+n2 
calculator(10,20)
```

```python
lambda n1,n2: n1 + n2
```

```python
# 삼각형, 사각형,원의 넓이 lambda함수
triArea = lambda h , w : h * w /2
recArea = lambda h, w: h * w
cirArea = lambda r : r * r *3.14

hei = int(input())
wid = int(input())
rad = int(input())

print(f'삼각형넓이 : {triArea(hei,wid)}')
print(f'사각형넓이: {recArea(hei,wid)}')
print(f'원의넓이: {cirArea(rad)}')
```
